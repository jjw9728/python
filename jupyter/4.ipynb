```python
import folium as fl
import random

# 랜덤 색상 생성 함수
def generate_random_color_hex():
    return "#{:06x}".format(random.randint(0, 0xFFFFFF))

# 지도 생성
m = fl.Map(location=[37.5665, 126.9780], zoom_start=12)

# 예시 데이터프레임 (taxi 데이터)
import pandas as pd

data = pd.read_csv("택시승차대 현황.csv", encoding='cp949')
taxi = data.loc[:, ["위치명", "위도", "경도"]]

# taxi = pd.DataFrame(data)

# 마커와 원형 마커 스타일링
for index, row in taxi.iterrows():
    random_color = generate_random_color_hex()  # 랜덤 색상 생성

    # Taxi 마커 추가 (아이콘과 색상)
    fl.Marker(
        [row[1], row[2]], 
        popup=fl.Popup("Taxi", max_width=200, min_width=100),  # 팝업을 더 예쁘게 꾸미기
        tooltip="Taxi Location",
        icon=fl.Icon(color=random_color, icon="fas fa-solid fa-taxi", icon_color="white", prefix="fa", icon_size=(25, 25))  # 아이콘 크기와 색상
    ).add_to(m)

    # 원형 마커 스타일링
    fl.CircleMarker(
        [row[1], row[2]],
        radius=15,  # 원의 반지름
        color=random_color,  # 테두리 색상
        fill=True,  # 채우기 색상 활성화
        fill_color=random_color,  # 채우기 색상
        fill_opacity=0.6,  # 채우기 투명도
        popup=fl.Popup("<b>Taxi Location</b><br>Radius: 15", max_width=300),  # 팝업에 HTML 태그 사용
        tooltip="<i>Click for details</i>",  # 툴팁을 이탤릭체로 설정
    ).add_to(m)

m
```
