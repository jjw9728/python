```python
import folium as fl
import random

#    
def generate_random_color_hex():
    return "#{:06x}".format(random.randint(0, 0xFFFFFF))

#  
m = fl.Map(location=[37.5665, 126.9780], zoom_start=12)

#   (taxi )
import pandas as pd

data = pd.read_csv(" .csv", encoding='cp949')
taxi = data.loc[:, ["", "", ""]]

# taxi = pd.DataFrame(data)

#    
for index, row in taxi.iterrows():
    random_color = generate_random_color_hex()  #   

    # Taxi   ( )
    fl.Marker(
        [row[1], row[2]], 
        popup=fl.Popup("Taxi", max_width=200, min_width=100),  #    
        tooltip="Taxi Location",
        icon=fl.Icon(color=random_color, icon="fas fa-solid fa-taxi", icon_color="white", prefix="fa", icon_size=(25, 25))  #   
    ).add_to(m)

    #   
    fl.CircleMarker(
        [row[1], row[2]],
        radius=15,  #  
        color=random_color,  #  
        fill=True,  #   
        fill_color=random_color,  #  
        fill_opacity=0.6,  #  
        popup=fl.Popup("<b>Taxi Location</b><br>Radius: 15", max_width=300),  #  HTML  
        tooltip="<i>Click for details</i>",  #   
    ).add_to(m)

m
```
